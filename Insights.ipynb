```python
from transformers import pipeline

# Load the "text2text-generation" pipeline with the "summarization" model
summarizer = pipeline("text2text-generation", model="t5-small", tokenizer="t5-small")

# Define a review text
review_text = df['review'][1]

# Define a prompt to generate a summary of the customer problem
prompt = f"'{review_text}'"

# Generate a summary of the customer problem using the T5 model
summary = summarizer(prompt)[0]['generated_text']

# Print the summary
print(summary)
```

```python
df['Summary'] = np.nan
```

```python
def get_review_summary(review_text):
  prompt = f"'{review_text}'"
  summary = summarizer(prompt)[0]['generated_text']
  return summary
```

```python
df['Summary'] = df['review'].apply(lambda s: get_review_summary(s) )
```

```python
df
```

```python
Verified_df_withSummary = df[(df['Verified_or_not'] == l[0]) | (df['Verified_or_not'] == l[1])]
```

```python
Verified_df_withSummary
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

# Split each review into tokens
tokenized_reviews = []
for review in Verified_df_withSummary['Summary']:
    tokens = word_tokenize(review)
    tokenized_reviews.extend(tokens)

# Print the tokenized reviews
print(tokenized_reviews)
```

```python
len(tokenized_reviews)
```

```python
import nltk
from nltk.corpus import stopwords

# Load the English stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Remove the stopwords from the list of words
filtered_word_list = [word for word in tokenized_reviews if word.lower() not in stop_words]

# Print the filtered list of words
print(filtered_word_list)
```

```python
len(filtered_word_list)
```

```python
cleaned_list = [''.join(c for c in s if c.isalpha()) for s in filtered_word_list]
```

```python
print(cleaned_list)
```

```python
len(cleaned_list)
```

```python
final_list = []
for i in cleaned_list:
  if i != '':
    final_list.append(i)
```

```python
len(final_list)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a WordCloud object
wordcloud = WordCloud(width=1000, height=500, background_color='white').generate(' '.join(final_list))

# Display the WordCloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from textblob import TextBlob

# Perform sentiment analysis using TextBlob
Verified_df_withSummary['polarity'] = Verified_df_withSummary['review'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Visualize the results using a histogram
import matplotlib.pyplot as plt
plt.hist(Verified_df_withSummary['polarity'])
plt.title('Sentiment Analysis of Customer Reviews')
plt.xlabel('Polarity')
plt.ylabel('Frequency')
plt.show()
```

```python
reviews_df = Verified_df_withSummary[:]
```

```python
#pip install vaderSentiment
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


# Preprocess the text data by removing punctuation and converting to lowercase
reviews_df['cleaned_review'] = reviews_df['review'].str.lower().replace('[^\w\s]','')

# Apply VADER sentiment analysis
analyzer = SentimentIntensityAnalyzer()
reviews_df['vader_scores'] = reviews_df['cleaned_review'].apply(lambda x: analyzer.polarity_scores(x))

# Visualize the results using a scatter plot
import matplotlib.pyplot as plt
plt.scatter(range(len(reviews_df)), reviews_df['vader_scores'].apply(lambda x: x['compound']))
plt.title('Sentiment Analysis of Customer Reviews')
plt.xlabel('Review')
plt.ylabel('VADER Sentiment Score')
plt.show()
```
