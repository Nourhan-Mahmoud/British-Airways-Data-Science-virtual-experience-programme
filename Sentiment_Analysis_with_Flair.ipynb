# Sentiment Analysis

```python
#pip install flair
```

```python
from flair.models import TextClassifier
from flair.data import Sentence
```

```python
classifier = TextClassifier.load('en-sentiment')
sentence = Sentence(df['review'][1])
classifier.predict(sentence)

# print sentence with predicted labels
print('Sentence above is: ', sentence.labels)
```

```python
sentence.labels[0].value
```

```python
sentence.labels[0].score
```

```python
df['Pos_or_Neg'] = np.nan
```

```python
df['Score'] = np.nan
```

```python
df
```

```python
# Define a function to get Flair sentiment prediction score
def score_flair(text):
  # Flair tokenization
  sentence = Sentence(text)
  # Predict sentiment
  classifier.predict(sentence)
  # Extract the score
  score = sentence.labels[0].score
  # Extract the predicted label
  value = sentence.labels[0].value
  # Return the score and the predicted label
  return score, value
```

```python
# Get sentiment score for each review
df['Score'] = df['review'].apply(lambda s: score_flair(s)[0])

# Predict sentiment label for each review
df['Pos_or_Neg'] = df['review'].apply(lambda s: score_flair(s)[1])
```
